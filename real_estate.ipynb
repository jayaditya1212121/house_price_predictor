```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
df1= df.drop(['availability','society'], axis= 1)
df1[0:3]
```

```python
df1.isnull().sum()
```

```python
df1.dropna(inplace = True)
df1.isnull().sum()
```

```python
df1['size'].unique()
```

```python
def x_bhk(x):
    return int(x.split(' ')[0])

x_bhk('4 bedroom')    
```

```python
df1['bhk']= df1['size'].apply(x_bhk)
df1[0:3]
```

```python
df_bhk=df1['size'].replace('[a-zA-Z]', '', regex = True, inplace= False)
df_bhk= df_bhk.apply(lambda x: int(x))
df_bhk[0:3]
```

```python
df1['bhk_alt']= df_bhk
df1[0:4]
```

```python
df1['total_sqft'].unique()
```

```python
def isfloat(x):
    try:
        float(x)
    except:
        return False
    return True

```

```python
df1[~ df1['total_sqft'].apply(isfloat)][0:20]
```

```python
def sqft_to_num(x):
    lst= x.split('-')
    if len(lst) == 2:
        return (float(lst[0])+float(lst[1]))/2
    try:
        return float(x)
    except:
        return None
    
```

```python
df1['total_sqft']= df1['total_sqft'].apply(sqft_to_num)
```

```python
df1.reset_index(inplace= True)
```

```python
df1.drop(['area_type', 'index'], axis =1, inplace= True)
```

```python
df1
```

```python
#outliers removed as per area/no of  bedrooms 
df2= df1[df1.total_sqft/df1.bhk_alt<300]
df2
```

```python
df1= df1[df1['total_sqft']/df1['bhk_alt']>400]
```

```python
df1['price_per_sqft']= df1['price']*100000/df1['total_sqft']
df1
```

```python
location_stats = df1.groupby('location')['location'].agg('count').sort_values(ascending = False)
location_stats
len(location_stats)
```

```python
location_remove= location_stats[location_stats<15]
len(location_remove)
```

```python
df1.location = df1.location.apply(lambda x:'other' if x in location_remove else x )
df1
```

```python
def remove_outliers(df):
    df_out= pd.DataFrame()
    for key, group in df1.groupby('location'):
        m1 = group.price_per_sqft.mean()
        std1= group.price_per_sqft.std()
        #new_group = group[(group.price_per_sqft<m1 + std1)&(group.price_per_sqft>m1 - std1)]
        new_group = group[(group.price_per_sqft<group.price_per_sqft.quantile(0.80))&(group.price_per_sqft>group.price_per_sqft.quantile(0.2))]
        df_out= pd.concat([df_out, new_group], ignore_index = True)
    return df_out
        
```

```python
df3 = remove_outliers(df1)
df3.price_per_sqft.describe()
```

```python
df3
```

```python
df4= df3[df3.bath < df3.bhk+2]
df4
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk_alt'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index)
    return df.drop(exclude_indices,axis='index')
df5= remove_bhk_outliers(df4)
```

```python
df5.drop(['size','price_per_sqft','bhk'], axis= 1, inplace= True)
df5
```

```python
df5.to_csv('df_clean.csv', index= False)
```

```python
df6=pd.get_dummies(df5.location)
df6.drop('other', axis = 1, inplace= True)
df6
```

```python
df7= pd.concat([df5,df6], axis = 1)
df7.drop('location', axis=1,inplace= True)
df7
```

```python
x= df7.drop(['price'], axis = 1)
x
```

```python
y= df7.price
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

x_train, x_test, y_train, y_test= train_test_split(x,y, test_size= 0.1, random_state= 1)
```

```python
model_lr = LinearRegression()
model_lr.fit(x_train,y_train)
model_lr.score(x_test,y_test)
```

```python
model_rr= Ridge()
model_rr.fit(x_train,y_train)
model_rr.score(x_test,y_test)
```

```python
model_lsr = Lasso()
model_lsr.fit(x_train,y_train)
model_lsr.score(x_test,y_test)
```

```python
def predict_price(location,sqft,bath,balcony,bhk):    
    loc_index = np.where(x.columns==location)[0][0]

    xp = np.zeros(len(x.columns))
    xp[0] = sqft
    xp[1] = bath
    xp[2] = balcony
    xp[3] = bhk
    if loc_index >= 4:
        xp[loc_index] = 1

    return model_lr.predict([xp])[0]
```

```python
predict_price('1st Phase JP Nagar', 2000,3,2,3)
```
